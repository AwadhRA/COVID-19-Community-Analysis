```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('US Census, Affected Communities.csv')
df["Median Household income( 2018 dollars)"] = df["Median Household income( 2018 dollars)"].str.replace(",","").astype(float)
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
from pandas import DataFrame

df_2 = DataFrame(df,columns=['Median Household income( 2018 dollars)','Percentage of minorities in population'])
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, init='k-means++').fit(df_2)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df_2['Median Household income( 2018 dollars)'], df_2['Percentage of minorities in population'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```python
from sklearn import metrics
n_clusters = [2,3,4,5,6,7,8]

for k in n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=49).fit(df_2)
    cluster_labels = kmeans.predict(df_2)
    S = metrics.silhouette_score(df_2, cluster_labels)
    print("n_clusters = {:d}, silhouette score {:1f}".format(k, S))
```

```python
cluster_map = pd.DataFrame()
cluster_map['data_index'] = df.index.values
cluster_map['cluster'] = kmeans.labels_
```

```python
df['Cluster'] = cluster_map['cluster'].astype(int)

df.head()
```
